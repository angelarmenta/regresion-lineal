```python
#developed by Roberto Ángel Meléndez-Armenta
#https://www.youtube.com/@educar-ia

#Fig 6
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Datos de ejemplo
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 3, 2, 5, 4])

# Crear el modelo y ajustarlo
model = LinearRegression()
model.fit(X, y)

# Predicciones
y_pred = model.predict(X)

# Calcular el coeficiente de determinación R²
r2 = r2_score(y, y_pred)

# Gráfico de dispersión con la línea de regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label=f'Línea de regresión (R² = {r2:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Coeficientes de Correlación y Determinación')
plt.legend()
plt.show()

```
