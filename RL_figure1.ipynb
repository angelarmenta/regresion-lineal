```python
#developed by Roberto Ángel Meléndez-Armenta
#https://www.youtube.com/@educar-ia

#Fig 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos de ejemplo
x = np.random.rand(50)
y = 2.5 * x + np.random.normal(size=x.size)

# Cálculo de la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Gráfico de dispersión con la línea de regresión
plt.scatter(x, y, label='Datos')
plt.plot(x, slope * x + intercept, color='red', label='Línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```
