```python
#Fig 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos de ejemplo
x = np.random.rand(50)
y = 2.5 * x + np.random.normal(size=x.size)

# Cálculo de la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Gráfico de dispersión con la línea de regresión
plt.scatter(x, y, label='Datos')
plt.plot(x, slope * x + intercept, color='red', label='Línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```

```python
#Fig 2
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo para diferentes correlaciones
x = np.random.rand(50)
y_pos = x + np.random.randn(50) * 0.1
y_neg = -x + np.random.randn(50) * 0.1
y_nulo = np.random.randn(50)

# Crear subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Correlación positiva
axs[0].scatter(x, y_pos, color='blue')
axs[0].set_title('Correlación Positiva')

# Correlación negativa
axs[1].scatter(x, y_neg, color='red')
axs[1].set_title('Correlación Negativa')

# Correlación nula
axs[2].scatter(x, y_nulo, color='green')
axs[2].set_title('Correlación Nula')

plt.show()
```

```python
#Fig 4
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 3, 2, 5, 4])

# Crear el modelo y ajustarlo
model = LinearRegression()
model.fit(X, y)

# Predicciones
y_pred = model.predict(X)

# Imprimir resultados
print(f'β1: {model.coef_[0]}')
print(f'β0: {model.intercept_}')

# Gráfico de dispersión con la línea de regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```

```python
#Fig 5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Datos de ejemplo
x = np.random.rand(50)
y = 2.5 * x + np.random.normal(size=x.size)

# Cálculo de la correlación
r, p_value = pearsonr(x, y)

# Imprimir resultados de la correlación
print(f'Coeficiente de correlación (r): {r}')
print(f'Valor p: {p_value}')
```
