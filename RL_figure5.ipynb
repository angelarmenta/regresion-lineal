```python
#developed by Roberto Ángel Meléndez-Armenta
#https://www.youtube.com/@educar-ia

#Fig 5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import linregress

# Datos de ejemplo
x = np.random.rand(50)
y = 2.5 * x + np.random.normal(size=x.size)

# Cálculo de la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Imprimir resultados
print(f'β1: {slope}')
print(f'β0: {intercept}')

# Cálculo de la correlación
r, p_value = pearsonr(x, y)

# Imprimir resultados de la correlación
print(f'Coeficiente de correlación (r): {r}')
print(f'Valor p: {p_value}')

# Gráfico de dispersión con la línea de regresión
plt.scatter(x, y, label='Datos reales')
plt.plot(x, slope * x + intercept, color='red', label='Línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```
