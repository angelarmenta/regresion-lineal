```python
#developed by Roberto Ángel Meléndez-Armenta
#https://www.youtube.com/@educar-ia

#Fig 2
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo para diferentes correlaciones
x = np.random.rand(50)
y_pos = x + np.random.randn(50) * 0.1
y_neg = -x + np.random.randn(50) * 0.1
y_nulo = np.random.randn(50)

# Crear subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Correlación positiva
axs[0].scatter(x, y_pos, color='blue')
axs[0].set_title('Correlación Positiva')

# Correlación negativa
axs[1].scatter(x, y_neg, color='red')
axs[1].set_title('Correlación Negativa')

# Correlación nula
axs[2].scatter(x, y_nulo, color='green')
axs[2].set_title('Correlación Nula')

plt.show()
```
